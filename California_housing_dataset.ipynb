```python
# 1- Importing datasets and important library required
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# 2- Fetching data from the sklearn's built-in dataset
data = fetch_california_housing ()
df = pd.DataFrame(
    data.data,
    columns=data.feature_names
)
df["target"] = data.target
print(df.head().to_string())
```

```python
# 3- Printing first 10 rows
print(df.head(10))
```

```python
# 4- Overall info
print(df.info())
```

```python
# 5- Splitting feature and target
x = data.data
y = data.target
print("Features--->")
print(x)
print("Target--->")
print(y)
```

```python
# 6- Checking for null values
print(df.isnull().sum())
```

```python
# 7- Splitting training and testing  (70%---> Training  30%---> Testing)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
        x,
        y,
        test_size = 0.3,
        random_state = 42
)
```

```python
# 8- Encoding
print(df.dtypes)


# Since every data is numeircal no need to do encoding
```

```python
# 9- Scaling data
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit_transform(X_train)
scale.fit(X_test)
```

```python
# 10- Model creation
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# 11- Making prediction
y_pred = model.predict(X_test)
print(y_pred[:5])
```

```python
# 12- Converting the 1D predited data to 2D for table like comparison
y_pred_2d = y_pred.reshape(len(y_pred),1)
y_pred_2d
```

```python
# 13- Converting the 1D y_test data to 2D for table like comparison
y_test_2d = y_test.reshape(len(y_test),1)
y_test_2d
```

```python
# 14- Creating dataframes of both 2D data with a defined heading
df1 = pd.DataFrame(y_pred_2d,columns=["Predicted"])
df2 = pd.DataFrame(y_test_2d, columns=['Real'])
```

```python
# 15- Concatinating both dataframes for getting a table-like comparison
pd.concat([df1,df2],axis=1)
```

```python
# 16- Concatinating both dataframes for getting a table-like comparison
comparison = pd.concat([df1,df2],axis=1)
print(comparison)
```

```python
# 17- Final statistics output
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print("MAE:", mean_absolute_error(y_test, y_pred))

print("MSE:", mean_squared_error(y_test, y_pred))

print("RMSE:", mean_squared_error(y_test, y_pred) ** 0.5)

print("R2 Score:", r2_score(y_test, y_pred))
```
